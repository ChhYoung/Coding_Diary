## 数据清洗
过滤一些统计有误的部分
- 血压： 舒张压高于压缩压得样本
- 身高： 低于2.5%和高于97.5%的样本
- 体重： 低于2.5%和高于97.5%的样本

```python
filtered_df =df[(df['ap_lo']<=df['ap_hi']) &
                (df['height'] >= df['height'].quantile(0.025))&
                (df['height'] <= df['height'].quantile(0.975))&
                (df['weight'] >= df['weight'].quantile(0.025))&
                (df['weight'] <= df['weight'].quantile(0.975))]
```

清洗掉的数据占的百分比

```python
1 - filtered_df.shape[0]/df.shape[0]
```

## 数据可视化

使用 `heatmap`绘制皮尔斯相关系数矩阵

```python
df = filtered_df.copy()
corr = df.corr(method='pearson')

#隐藏上三角,获取上三角mask
mask = np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

# 绘制图像
f,ax=plt.subplots(figsize=(12,9))
sns.heatmap(corr,mask=mask,vmax=1,center=0,annot=True,fmt='.1f',
           square=True,linewidths=.5,cbar_kws={"shrink":.5})
```

## 男女身高分布
1. 绘制身高和性别之间的小提琴图 `violinplot()`

通过`hue`参数按性别分类，`scale`参数来计算性别对应的具体数量

```python
df_melt = pd.melt(frame=df, value_vars=['height'],id_vars=['gender'])
plt.figure(figsize=(12,10))
ax=sns.violinplot(
    x='variable',
    y='value',
    hue='gender',
    palette="muted",
    split=True,
    data=df_melt,
    scale='count',
    scale_hue = False
)
```

绘制身高和性别之间的和密度图,可以通过和密度图去除地看到性别之间地差异，但缺无法得到每个性别对应地具体人数

```python
sns.FacetGrid(df,hue="gender",height=12)\
    .map(sns.kdeplot,"height").add_legend()
```

Spearman's rank correlation coefficient 斯皮尔曼相关系数，利用单调方程评价两个统计变量地相关性，用于衡量两个变量地依赖性地非参数指标

```python
corr=df[['id','age','height','weight','ap_hi','ap_lo','cholesterol','gluc']].corr(method='spearman')

mask = np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

f,ax=plt.subplots(figsize=(12,10))

sns.heatmap(corr,mask=mask,vmax=1,center=0,annot=True,fmt='.2f',
           square=True,linewidths=.5,cbar_kws={"shrink":.5})
```
